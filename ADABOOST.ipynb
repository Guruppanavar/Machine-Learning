# ADABOOST
- AdaBoost can be used to boost the performance of any machine learning algorithm. It is best used with weak learners. These are models that achieve accuracy just above random chance on a classification problem.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
hr = pd.read_csv('HR Analytics.csv')
hr.head()
```

```python
new_hr = hr.drop('Attrition',axis=1)
new_hr
```

```python
hr_dummies = pd.get_dummies(new_hr)
hr_dummies['Attrition'] = hr['Attrition']
hr_dummies.head()
```

```python
train , test = train_test_split(hr_dummies, test_size=0.3, random_state=100)
from sklearn.metrics import confusion_matrix
```

```python
train_y = train['Attrition']
test_y= test['Attrition']
train_x= train.drop('Attrition', axis=1) # for row wise operation using col
test_x= test.drop('Attrition',axis=1)
```

```python
model= AdaBoostClassifier(random_state=100)
model.fit(train_x,train_y)
test_pred = model. predict(test_x)
tn,fp,fn,tp = confusion_matrix(test_y,test_pred).ravel()
print(tn)
print(fp)
print(fn)
print(tp)
```

```python
accuracy= (tp+tn)/(tp+tn+fp+fn)
accuracy
```

```python
sensitivity=tp/(tp+fn)
sensitivity
```

```python
specificity=tn/(fp+tn)
specificity
```
